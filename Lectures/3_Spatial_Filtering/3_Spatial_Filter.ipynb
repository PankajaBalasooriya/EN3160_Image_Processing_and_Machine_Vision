```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Spatial Filtering

To Enhance images (Noise Filtering), Locate Objects (Template Matching) spatial filtering is used. 

In linear spatial filtring we usee 2-D comvolution operation. 

### Using Padding
When dealing wth image boundaries we need to use padding to control the size of the output. 

Folloiwng value can be used to pad the image:
    1. Zero pad
    2. Wrap around
    3. Copy edge
    4. Reflect across edge

## Averaging

```python
sigiriya_img = cv.imread("./images/Sigiriya.jpeg", cv.IMREAD_GRAYSCALE)
assert sigiriya_img is not None

kernal = np.ones((3, 3),np.float32)/9

sigiriya_img_avg = cv.filter2D(sigiriya_img, cv.CV_32F, kernal)

fig , axes = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(18, 9))

axes[0].imshow(sigiriya_img, cmap='grey')
axes[0].set_title('Original')
axes[0].set_xticks([]), axes[0].set_yticks([])

axes[1].imshow(sigiriya_img_avg, cmap='grey')
axes[1].set_title('Averaging')
axes[1].set_xticks([]), axes[1].set_yticks([])

plt.show()
```

# Sobel Filtering

```python
sobel_x = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

sobel_y = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sigiriya_x = cv.filter2D(sigiriya_img, cv.CV_64F, sobel_x)
sigiriya_y = cv.filter2D(sigiriya_img, cv.CV_64F, sobel_y)

fig , axes = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(18, 9))

axes[0].imshow(sigiriya_x, cmap='grey')
axes[0].set_title('Sobel X')
axes[0].set_xticks([]), axes[0].set_yticks([])

axes[1].imshow(sigiriya_y, cmap='grey')
axes[1].set_title('Sobel Y')
axes[1].set_xticks([]), axes[1].set_yticks([])
```
